# Hypothesis 1 — Does the Test group have a higher completion rate?

### Step 1 — Define the hypothesis
Here we write the two statements we want to test.

- **Null hypothesis (H0):** The Test group does **not** have a higher completion rate than the Control group.
- **Alternative hypothesis (H1):** The Test group **does** have a higher completion rate than the Control group.

```python
# We write the hypothesis as simple text so it is easy to present
H0_1 = 'The Test group does not have a higher completion rate than the Control group.'
H1_1 = 'The Test group has a higher completion rate than the Control group.'

print('H0:', H0_1)
print('H1:', H1_1)
```

### Step 2 — Separate the two groups
We first split the data into the Control group and the Test group so we can compare them.

```python
control_group = ab_data[ab_data[variation_col] == 'Control'].copy()
test_group = ab_data[ab_data[variation_col] == 'Test'].copy()

print('Control group size:', control_group.shape[0])
print('Test group size:', test_group.shape[0])
```

### Step 3 — Calculate the completion rate for each group
A completion rate is just:

**number of completed clients / total number of clients**

This tells us the share of clients that finished the process.

```python
control_rate = control_group['completed'].mean()
test_rate = test_group['completed'].mean()
difference_rate = test_rate - control_rate

print('Control completion rate:', round(control_rate, 4))
print('Test completion rate:', round(test_rate, 4))
print('Difference:', round(difference_rate, 4))
```

### Step 4 — Count the successes and totals
For a proportion test, we need two things:
- how many people completed the process
- how many people were in each group

In this case, a **success** means a client completed the process.

```python
control_success = control_group['completed'].sum()
test_success = test_group['completed'].sum()

control_total = control_group.shape[0]
test_total = test_group.shape[0]

print('Control successes:', control_success)
print('Test successes:', test_success)
print('Control total:', control_total)
print('Test total:', test_total)
```

### Step 5 — Choose the test and set alpha
We use a **proportion z-test** because we are comparing **two percentages**.

We also set **alpha = 0.05**. This means we allow a 5% chance of being wrong.

```python
alpha = 0.05
print('Chosen test: Proportion z-test')
print('Alpha:', alpha)
```

### Step 6 — Run the statistical test
Now we run the test. It gives us two main results:
- **z-statistic**
- **p-value**

The p-value is the most important one for our decision.

```python
z_stat_1, p_value_1 = proportions_ztest(
    count=[test_success, control_success],
    nobs=[test_total, control_total],
    alternative='larger'
)

print('Z-statistic:', round(z_stat_1, 4))
print('P-value:', p_value_1)
```

### Step 7 — Make the decision
Decision rule:
- if p-value < 0.05, we reject H0
- if p-value >= 0.05, we fail to reject H0

This tells us whether the difference is likely real or just due to chance.

```python
if p_value_1 < alpha:
    decision_1 = 'Reject H0'
    simple_result_1 = 'The Test group has a statistically significant higher completion rate.'
else:
    decision_1 = 'Fail to reject H0'
    simple_result_1 = 'We do not have enough evidence to say the Test group is better.'

print('Decision:', decision_1)
print('Simple result:', simple_result_1)
```

### Step 8 — Write the business explanation
This final step turns the statistical result into very simple English for presentation.

```python
print('Business explanation:')
print('The new design increased completion rate from', round(control_rate * 100, 2), '% to', round(test_rate * 100, 2), '%.')
print('The uplift is', round(difference_rate * 100, 2), 'percentage points.')
print(simple_result_1)
```

# Hypothesis 2 — Is the improvement at least 5%?

### Step 1 — Define the hypothesis
Vanguard said the new design must improve completion by **at least 5%** to be worth the cost.

- **Null hypothesis (H0):** The improvement is **less than 5%**
- **Alternative hypothesis (H1):** The improvement is **at least 5%**

```python
H0_2 = 'The improvement is less than 5%.'
H1_2 = 'The improvement is at least 5%.'

print('H0:', H0_2)
print('H1:', H1_2)
```

### Step 2 — Show the completion rates again
We reuse the completion rates from Hypothesis 1 because the 5% rule is based on the same rates.

```python
print('Control completion rate:', round(control_rate * 100, 2), '%')
print('Test completion rate:', round(test_rate * 100, 2), '%')
```

### Step 3 — Calculate the observed uplift
The **uplift** means how much better the Test group performed than the Control group.

```python
observed_uplift = test_rate - control_rate
required_uplift = 0.05

print('Observed uplift:', round(observed_uplift * 100, 2), '%')
print('Required uplift:', round(required_uplift * 100, 2), '%')
```

### Step 4 — Compare the observed uplift to the threshold
This is the most direct way to answer the cost-effectiveness question.

```python
difference_from_threshold = observed_uplift - required_uplift

print('Observed uplift minus required uplift:', round(difference_from_threshold * 100, 2), '%')
```

### Step 5 — Explain why this is a business test
This is different from the first hypothesis.

Hypothesis 1 asks: **Is the result real?**
Hypothesis 2 asks: **Is the result big enough to matter for the business?**

```python
print('This step is about business value, not just statistical significance.')
print('We want to know if the redesign is strong enough to justify its cost.')
```

### Step 6 — Make the threshold decision
We compare the uplift directly with 5%.

```python
if observed_uplift >= required_uplift:
    decision_2 = 'Reject H0'
    simple_result_2 = 'The redesign meets Vanguard's 5% threshold.'
else:
    decision_2 = 'Fail to reject H0'
    simple_result_2 = 'The redesign does not meet Vanguard's 5% threshold.'

print('Decision:', decision_2)
print('Simple result:', simple_result_2)
```

### Step 7 — Connect it with Hypothesis 1
This step is useful in a presentation because it shows both sides of the story: the redesign can be statistically significant but still not large enough for business needs.

```python
print('Hypothesis 1 result:', simple_result_1)
print('Hypothesis 2 result:', simple_result_2)
```

### Step 8 — Write the final business conclusion
Now we write the plain-English business conclusion.

```python
print('Final business conclusion:')
print('The redesign improved completion by', round(observed_uplift * 100, 2), '%.')
print('Vanguard needed at least 5.00%.')
print(simple_result_2)
```

# Hypothesis 3 — Is the average age the same in the Test and Control groups?

### Step 1 — Define the hypothesis
This is our extra hypothesis.

- **Null hypothesis (H0):** The average age is the same in both groups.
- **Alternative hypothesis (H1):** The average age is different in the two groups.

```python
H0_3 = 'The average age is the same in the Test and Control groups.'
H1_3 = 'The average age is different in the Test and Control groups.'

print('H0:', H0_3)
print('H1:', H1_3)
```

### Step 2 — Merge the A/B data with demographics
To compare ages, we need to bring the age column into our A/B testing table.

```python
age_data = ab_data.merge(df_demo[['client_id', 'clnt_age']], on='client_id', how='left')

print(age_data[['client_id', variation_col, 'completed', 'clnt_age']].head())
```

### Step 3 — Remove missing ages
We remove rows where age is missing so the test uses clean data.

```python
age_data = age_data.dropna(subset=['clnt_age']).copy()

print('Rows after removing missing ages:', age_data.shape[0])
```

### Step 4 — Separate ages into Test and Control
Now we create one age list for the Test group and one for the Control group.

```python
control_age = age_data[age_data[variation_col] == 'Control']['clnt_age']
test_age = age_data[age_data[variation_col] == 'Test']['clnt_age']

print('Control age count:', control_age.shape[0])
print('Test age count:', test_age.shape[0])
```

### Step 5 — Calculate the average age for each group
Before running the test, it is good to look at the average age itself.

```python
control_age_mean = control_age.mean()
test_age_mean = test_age.mean()
age_difference = test_age_mean - control_age_mean

print('Control average age:', round(control_age_mean, 2))
print('Test average age:', round(test_age_mean, 2))
print('Difference in average age:', round(age_difference, 2))
```

### Step 6 — Choose the test and run it
We use a **t-test** here because we are comparing **two averages**.

We use `equal_var=False` to keep it simple and safe when the two groups may not have exactly the same spread.

```python
t_stat_3, p_value_3 = ttest_ind(test_age, control_age, equal_var=False)

print('T-statistic:', round(t_stat_3, 4))
print('P-value:', p_value_3)
```

### Step 7 — Make the decision
Again we compare the p-value to 0.05.

```python
if p_value_3 < alpha:
    decision_3 = 'Reject H0'
    simple_result_3 = 'There is a statistically significant age difference between the groups.'
else:
    decision_3 = 'Fail to reject H0'
    simple_result_3 = 'There is no statistically significant age difference between the groups.'

print('Decision:', decision_3)
print('Simple result:', simple_result_3)
```

### Step 8 — Write the practical explanation
This last step helps us explain the result in simple terms.

Even if a difference is statistically significant, it may still be very small in real life.

```python
print('Practical explanation:')
print('The average age difference is', round(age_difference, 2), 'years.')
print(simple_result_3)
print('So we should look at whether this difference is actually important in practice, not only statistically.')
```
